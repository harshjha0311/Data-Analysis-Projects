# EXPLAINATORY DATA ANALYSIS

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df, hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values( by='Amount', ascending= False )

sales_gen_graph=sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender')

for label in sales_gen_graph.containers:
    sales_gen_graph.bar_label(label)
```

### Age

```python
age_range= sns.countplot( x='Age Group', data=df, hue='Gender') 

for bars in age_range.containers:
    age_range.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)
sales_age_graph=sns.barplot(x= 'Age Group', y='Amount', data=sales_age, hue='Age Group')
```

### State


```python
state_order= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
state_order_gr=sns.barplot( x='State', y='Orders', data=state_order, hue='State')

for bar in state_order_gr.containers:
    state_order_gr.bar_label(bar)
```

```python
state_amount= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(state_amount)
```

```python
state_amount= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
state_amount_gr=sns.barplot( x='State', y='Amount', data=state_amount, hue='State')

for bar in state_amount_gr.containers:
    state_amount_gr.bar_label(bar)
```

### Marital Status

```python
ms_count=df.groupby(['Marital_Status'])['Marital_Status'].count()
ms_order_gr= plt.pie(ms_count, labels=['Married', 'Unmarried'], colors=sns.color_palette("bright"), autopct='%.0f%%')
```

```python
ms_amount=df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ms_amount_gr=plt.pie(ms_amount['Amount'], labels=ms_amount['Marital_Status'], colors=sns.color_palette('bright'), autopct='%.0f%%')
```

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', hue='Gender', palette='bright')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue='Occupation')
```

## Zone

```python
zone_order=sns.countplot(x='Zone', data=df, hue='Gender', order=df['Zone'].value_counts().index)

for i in zone_order.containers:
    zone_order.bar_label(i)
```

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
plt.plot(sales_state['Product_ID'], sales_state['Orders'], color='black')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
